## tl;dr

ショート版イージーの初期能力を **15〜40から12〜37へ変更**する案を採用する。保存済みの5,000シード評価では、S率が66.70%から51.22%、S+A率が85.68%から75.20%へ低下する。

## Context & Methods

128個の合法的な方針候補を、選定80シード、上位5方針の検証400シード、未使用5,000シードの順で評価した。最適化目標は平均ポイント。

### Key Assumptions

- ユーザーから目標比率の指定がまだないため、S率45〜55%、S+A率70〜80%を暫定目標とした。
- イージー専用の初期能力だけを変更し、練習・勉強倍率や大会難度は固定した。
- 数学的完全解ではなく、定義済み128方針の中での平均ポイント最良候補である。

## Data

```python
import json
from pathlib import Path

source_candidates = [
    Path('docs/analysis/2026-07-23-easy-balance-sweep.json'),
    Path('2026-07-23-easy-balance-sweep.json'),
]
source_path = next(path for path in source_candidates if path.exists())
data = json.loads(source_path.read_text(encoding='utf-8'))
scenarios = data['scenarios']
len(scenarios), data['recommendedScenario']
```

## Results

```python
header = f"{'scenario':<20} {'range':<7} {'S':>8} {'S+A':>8} {'mean pt':>10}"
print(header)
print('-' * len(header))
for row in scenarios:
    start_range = f"{row['statMin']}-{row['statMax']}"
    print(f"{row['id']:<20} {start_range:<7} {row['sRate']:>7.2%} {row['saRate']:>7.2%} {row['meanPoints']:>10.1f}")
```

```python
for row in scenarios:
    rank_total = sum(row[key] for key in ['sRate', 'aRate', 'bRate', 'cRate', 'dRate', 'eRate', 'expelledRate'])
    assert abs(rank_total - 1) < 1e-9, (row['id'], rank_total)

recommended = next(row for row in scenarios if row['id'] == data['recommendedScenario'])
target = data['provisionalTarget']
assert target['sRateMin'] <= recommended['sRate'] <= target['sRateMax']
assert target['saRateMin'] <= recommended['saRate'] <= target['saRateMax']
recommended
```

## Takeaways

- `statMin: 12, statSpread: 26` は暫定目標内に入る。
- `statMin: 15, statSpread: 20〜21` でも近い結果になるが、開始時の振れ幅が狭くなる。
- 下限だけを変える案は周回ごとの初期キャラクター差を維持でき、変更箇所も1パラメータで済む。
- 次の判断材料は、人間プレイでイージーが本当に『少し失敗はするが気持ちよく勝てる』範囲かどうか。
